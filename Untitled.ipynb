```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def BDC(x, d):
    y = []
    for i in range(x.shape[0]):
        if np.random.random() < 1 - d:
            y.append(x[i])
    return np.array(y, dtype=bool)

def sample_ber_d(n, d):
    return np.random.random(n) < d

def entr(d):
    res = []
    EPS = 1e-10
    for v in d:
        if abs(v) < EPS or abs(1-v) < EPS:
            res.append(0)
        else:
            res.append(-v*np.log2(v) -(1-v)*np.log2(1-v))
    return np.array(res)
```

```python
# A Dynamic Programming based Python
# program to find the number of times
# the second string occurs in the first
# string, whether continuous or discontinuous
 
# Iterative DP function to find the 
# number of times the second string
# occurs in the first string,
# whether continuous or discontinuous
def count(a, b):
    m = len(a)
    n = len(b)
 
    # Create a table to store results of sub-problems
    lookup = [[0] * (n + 1) for i in range(m + 1)]
 
    # If first string is empty
    for i in range(n+1):
        lookup[0][i] = 0
 
    # If second string is empty
    for i in range(m + 1):
        lookup[i][0] = 1
 
    # Fill lookup[][] in bottom up manner
    for i in range(1, m + 1):
        for j in range(1, n + 1):
             
            # If last characters are same, 
            # we have two options -
            # 1. consider last characters of 
            # both strings in solution
            # 2. ignore last character of first string
            if a[i - 1] == b[j - 1]:
                lookup[i][j] = lookup[i - 1][j - 1] + lookup[i - 1][j]
                 
            else:
                # If last character are different, ignore
                # last character of first string
                lookup[i][j] = lookup[i - 1][j]
 
    return lookup[m][n]
```

```python
def estimate_Einf(n, d, nsamples):
    sum_vals = 0
    for i in range(nsamples):
        X = sample_ber_d(n, 0.5)
        Y = BDC(X, d)
        c = count(X, Y)
        if c == 0:
            print(f'error with X={X}, Y={Y}')
        sum_vals += np.log2(float(c)) / n
    return sum_vals / nsamples
```

```python
d_vals = np.linspace(0, 1, 20)[1:-1]
for j in range(5, 9):
    print(f'At j={j}')
    n = 2**j
    E_inf_vals = []
    for d in d_vals:
        E_inf_vals.append(estimate_Einf(n, d, 100))
    plt.plot(d_vals, E_inf_vals, label=f'n={n}')
    plt.legend()
```

```python
plt.plot(d_vals, d_vals + entr(d_vals) -np.array(E_inf_vals))
plt.plot(d_vals, d_vals*0 + 1)
plt.axvline(x=0.5)
```
